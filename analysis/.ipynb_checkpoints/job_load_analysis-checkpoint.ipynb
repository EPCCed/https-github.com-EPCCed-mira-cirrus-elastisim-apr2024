```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import csv
import json
```

```python
csvfile = open('2024-Q1_Cirrus-CPU_moldable-simulated.csv', 'r')

csvreader = csv.DictReader(csvfile)
job_l = []
for row in csvreader:
    job_l.append(row)

print(job_l[0])
print(job_l[-1])
```

```python
jsonfile = open('cirrus-cpu-simple/data/input/2024_Q1_CPU-jobs.json', 'r')
jobs_d = json.load(jsonfile)
i = 0
maxtime = 0
totuse = 0
for job in jobs_d['jobs']:
    job_l[i]['Nodes'] = int(job['num_nodes'])
    totuse = totuse + int(job_l[i]['Makespan'])
    if int(job_l[i]['End Time']) > maxtime:
        maxtime = int(job_l[i]['End Time'])
    i += 1

print(len(job_l))
print(maxtime)
print(totuse)
```

```python
load_a = np.zeros(maxtime+1, dtype=int)
```

```python
for job in job_l:
    temp_a = np.zeros(maxtime+1, dtype=int)
    istart = int(job['Start Time'])
    iend = int(job['End Time'])
    nodes = int(job['Nodes'])
    temp_a[istart:iend] = nodes
    load_a = load_a + temp_a

print(load_a)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(load_a)
plt.xlabel("Time")
plt.ylabel("Cores in use")
sns.despine()
# plt.savefig("simulated_load.png")
```

```python
csvfile = open('raw_job_statistics.csv', 'r')

csvreader = csv.DictReader(csvfile)
job_l = []
maxtime = 0
for row in csvreader:
    if int(row['End Time']) > maxtime:
        maxtime = int(row['End Time'])
    if row['Nodes'] is None:
        print(row)
    job_l.append(row)

print(maxtime)
```

```python
load_a = np.zeros(maxtime+1, dtype=int)
```

```python
totuse = 0
for job in job_l:
    temp_a = np.zeros(maxtime+1, dtype=int)
    istart = int(job['Start Time'])
    iend = int(job['End Time'])
    nodes = int(job['Nodes'])
    totuse = totuse + int(job['Makespan'])
    temp_a[istart:iend] = nodes
    load_a = load_a + temp_a

print(totuse)
```

```python
plt.clf()
plt.figure(figsize=(15, 5))
plt.plot(load_a)
plt.xlabel("Time")
plt.ylabel("Cores in use")
sns.despine()
plt.savefig("measured_load.png")
```
